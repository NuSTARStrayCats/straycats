```python
from astropy.io.fits import getdata
from astropy.table import Table
#straycats = getdata('../straycats.fits')
st = Table.read('../straycats.fits')
for key in st.columns:
    if st[key].dtype.str.startswith('|S'):
        st[key] = st[key].astype('str') 
straycats = st.to_pandas()
```
